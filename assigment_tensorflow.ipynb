```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
```

```python
print(tf.__version__)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 1234)
```

```python
input_dim=X_train.shape[1]
```

```python
print(input_dim)
```

```python
# column order in CSV file
column_names = ['CreditScore','Geography','Gender''Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Exited']

feature_names = column_names[:-1]
label_name = column_names[-1]

print("Features: {}".format(feature_names))
print("Label: {}".format(label_name))
```

```python
# Create an Artificial Neural Network model
# JST models that are made using a hard framework with specifications:

# 2 hidden layers with 10 neurons and 10 neurons
#neuron uses the ReLU activation function
#and the output layer with the sigmoid activation function
```

```python
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#do learning
```

```python
model.fit(X_train, y_train, epochs=150, batch_size=10000)
```

```python
accuracy = model.evaluate(X_test, y_test)
```

```python
model.save('my_model.h5')
```
